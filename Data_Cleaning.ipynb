```python
measurements = pd.read_csv('c:/Users/Paolo/Documents/Bioinformatics Research/Khomtchouk Lab/Aachen University/MA_VC_Study/LABEVENTS.csv')
```

```python
measurements
```

```python
pH = measurements[measurements['ITEMID']==50820][['SUBJECT_ID','VALUENUM']]
pH = pH.drop_duplicates(subset=['SUBJECT_ID'], keep='first')
lactate = measurements[measurements['ITEMID']==50813][['SUBJECT_ID','VALUENUM']]
lactate = lactate.drop_duplicates(subset=['SUBJECT_ID'], keep='first')
bicarbonate = measurements[measurements['ITEMID']==50882][['SUBJECT_ID','VALUENUM']]
bicarbonate = bicarbonate.drop_duplicates(subset=['SUBJECT_ID'], keep='first')
base_excess = measurements[measurements['ITEMID']==50802][['SUBJECT_ID','VALUENUM']]
base_excess = base_excess.drop_duplicates(subset=['SUBJECT_ID'], keep='first')
pCO2 = measurements[measurements['ITEMID']==50818][['SUBJECT_ID','VALUENUM']]
pCO2 = pCO2.drop_duplicates(subset=['SUBJECT_ID'], keep='first')
sodium = measurements[measurements['ITEMID']==50983][['SUBJECT_ID','VALUENUM']]
sodium = sodium.drop_duplicates(subset=['SUBJECT_ID'], keep='first')
chloride = measurements[measurements['ITEMID']==50902][['SUBJECT_ID','VALUENUM']]
chloride = chloride.drop_duplicates(subset=['SUBJECT_ID'], keep='first')
```

```python
pH = pH.rename(columns={'VALUENUM': 'pH'})
lactate = lactate.rename(columns={'VALUENUM': 'lactate'})
bicarbonate = bicarbonate.rename(columns={'VALUENUM': 'bicarbonate'})
base_excess = base_excess.rename(columns={'VALUENUM': 'base_excess'})
pCO2 = pCO2.rename(columns={'VALUENUM': 'pCO2'})
sodium = sodium.rename(columns={'VALUENUM': 'sodium'})
chloride = chloride.rename(columns={'VALUENUM': 'chloride'})
```

```python
cohort = pH.merge(lactate, on='SUBJECT_ID')
cohort = cohort.merge(bicarbonate, on='SUBJECT_ID')
cohort = cohort.merge(base_excess, on='SUBJECT_ID')
cohort = cohort.merge(pCO2, on='SUBJECT_ID')
cohort = cohort.merge(sodium, on='SUBJECT_ID')
cohort = cohort.merge(chloride, on='SUBJECT_ID')
```

```python
cohort['VC'] = False
cohort.loc[cohort['SUBJECT_ID'].isin(subjects_vc_index), 'VC'] = True
cohort
```

```python
cohort.to_csv('MIMIC_III_VC.csv', index=False)
```
